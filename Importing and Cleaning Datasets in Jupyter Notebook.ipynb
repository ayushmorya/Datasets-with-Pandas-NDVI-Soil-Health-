```python
import pandas as pd
df = pd.read_csv('ndvi_soil_health_dataset.csv')
print(df.head())
```

**Inspecting the Dataset**

```python
#first check the issues with the dataset before we can use it

print(df.info())
print(df.describe())
```

**Cleaning the dataset**

Missing values, Outliers, incorrect data formats

```python
#checking missing values
print(df.isnull().sum())
```

```python
#Hndling Outliers
#Outliers like NDVI of 999 are errors since NDVI must be between -1 to 1

df['NDVI'] = df['NDVI'].clip(lower = -1, upper = 1)
```

```python
#Now convert Levels
#Our health column is text Healthy or Unhealthy but the models need numbers like O for Unhelathy and 1 for healthy

df['Health'] = df['Health'].map({'healthy':1, 'unhealthy': 0})
df.info()
```
